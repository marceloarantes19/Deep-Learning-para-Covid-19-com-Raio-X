```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
vet = np.zeros((50,6))
tit = []
pl = True
i = -1
with open('arquivos/AcuraciaComparativo.csv', 'r') as arq:
  l = csv.reader(arq)
  for linha in l:
    j = 0
    for x in linha:
      if pl:
        tit.append(x) 
      else:
        vet[i][j] = float(x)
        j = j + 1
    pl = False
    i = i + 1
vet = np.transpose(vet)
```

```python
plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, 50), vet[0], label="ShallowNet")
plt.plot(np.arange(0, 50), vet[1], label="LeNet")
plt.plot(np.arange(0, 50), vet[2], label="AlexNet")
plt.plot(np.arange(0, 50), vet[3], label="VGGNet")
plt.plot(np.arange(0, 50), vet[4], label="CXNet")
plt.plot(np.arange(0, 50), vet[5], label="CoviXNet")
plt.title("Acurcia")
plt.xlabel("pocas")
plt.ylabel("Valores")
plt.legend()
plt.show()
```
